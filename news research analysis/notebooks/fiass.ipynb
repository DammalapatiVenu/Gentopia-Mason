```python
import pandas as pd
pd.set_option('display.max_colwidth', 100)
```

```python
df = pd.read_csv("/content/drive/MyDrive/ExtraCredit/smpl.csv")
df.shape
```

```python
df
```

```python
from sentence_transformers import SentenceTransformer
```

```python
encdr = SentenceTransformer("all-mpnet-base-v2")
vctrs = encdr.encode(df.text)
```

```python
vctrs.shape
```

```python
dm = vctrs.shape[1]
dm
```

```python
import faiss
idx = faiss.IndexFlatL2(dm)
```

```python
idx.add(vctrs)
idx
```

```python
# search_query = "I want to buy a polo t-shirt"
search_query = "An apple a day keeps the doctor away"
vec = encdr.encode(search_query)
vec.shape
```

```python
vec
```

```python
import numpy as np
svec = np.array(vec).reshape(1,-1)
svec.shape
```

```python
idx.search(svec,k=2)
```

```python
dstncs, I = idx.search(svec, k=2)
dstncs
```
